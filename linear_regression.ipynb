```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns

from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#doing this to get general understanding of dataset
df = pd.read_csv("student_performance_large_dataset.csv")
df.head()
print(df.columns)
print(df.info())

#data preprocessing (step 1) -> identifying null values and the datatype of the data
print(df.isnull().sum())
print(df.describe)
```

```python
X_sleep = df[['Sleep_Hours_per_Night']] #used bcs to access the feature as 2D df it needs the double brackets -> data preprocessing 
y = df["Exam_Score (%)"]
 # data preprocessing -> making sure everything is in the correct format
print(X_sleep.shape)
print(y.shape)
x_train, x_test, y_train, y_test = train_test_split(X_sleep, y, test_size=0.2, random_state=8)
model_sleep = LinearRegression()
model_sleep.fit(x_train, y_train)
y_predict = model_sleep.predict(x_test)

print("sleep feature model: ")
print("R^2 score:", r2_score(y_test,y_predict))
print("MSE", mean_squared_error(y_test,y_predict))
print("slope(coefficient): ", model_sleep.coef_[0])
print("intercept: ", model_sleep.intercept_)

plt.scatter(x_test, y_test, color="blue", alpha=0.3 ,label="Actual Data") #doing the scatter plot
plt.plot(x_test, y_predict, color="red", label="Regression line") #plot regression line
plt.title("Sleep vs. Final Grade (Simple Linear Regression)")
plt.xlabel("Sleep Hours per Night")
plt.ylabel("Final Grade")
plt.legend()
plt.grid(True)
plt.show()
sort_idx = np.argsort(x_test.values.flatten())
plt.plot(x_test.values.flatten()[sort_idx], y_predict[sort_idx], color="red", label="Regression Line")


```

```python
X_all = df[['Study_Hours_per_Week', 'Time_Spent_on_Social_Media (hours/week)','Attendance_Rate (%)','Sleep_Hours_per_Night']] # used bcs to access the feature as 2D df it needs the double brackets
y = df["Exam_Score (%)"]
print(X_all.shape)
print(y.shape)
x_train, x_test, y_train, y_test = train_test_split(X_all, y, test_size=0.2, random_state=8)
model_all = LinearRegression()
model_all.fit(x_train, y_train)
y_predict = model_all.predict(x_test)

print(y_predict.shape)
print(y_test.shape)

print("multiple feature model: ")
print("R^2 score:", r2_score(y_test,y_predict))
print("MSE", mean_squared_error(y_test,y_predict))

coefficients = pd.DataFrame({
   'Feature': X_all.columns,
   'Coefficient': model_all.coef_
})
print(coefficients)



coefficients['abs_coef'] = coefficients['Coefficient'].abs()
coefficients_sorted = coefficients.sort_values(by='abs_coef', ascending=False)

coefficients_sorted = coefficients_sorted.drop(columns='abs_coef')
print(coefficients_sorted)

plt.figure(figsize=(10,6))
plt.barh(coefficients_sorted['Feature'], coefficients_sorted['Coefficient'], color='skyblue')
plt.xlabel('coefficient value')
plt.title('Feature Importance in Predicting Final Grade')
plt.axvline(
    x = 0,
    color= 'gray',
    linestyle = '--'
)
plt.gca().invert_yaxis()  #to make the highest impact at the top
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
Xarray_studyHoursPerWeek = X_studyHoursPerWeek.to_numpy()
sum = 0
for i in range(len(Xarray_studyHoursPerWeek)):
    sum += Xarray_studyHoursPerWeek[i]

avgX = sum / len(Xarray_studyHoursPerWeek)

diff_X = []
for i in range(len(Xarray_studyHoursPerWeek)):
    diff_X.append(Xarray_studyHoursPerWeek[i] - avgX)

diff_X = np.array(diff_X)

Yarray = y.to_numpy()
sum2 = 0
for i in range(len(y)):
    sum2 += Yarray[i]

avgY = sum2 / len(Yarray)

diff_Y = []
for i in range(len(Yarray)):
    diff_Y.append(Yarray[i] - avgY)

diff_Y = np.array(diff_Y)

sum1 = 0
sum2 = 0
sum3 = 0
for i in range(len(diff_X)):
    sum1 += diff_X[i]*diff_Y[i]
    sum2 += diff_X[i]*diff_X[i]
    sum3 += diff_Y[i]*diff_Y[i]

pearsonCoef = sum1 / (np.sqrt(sum2*sum3))
print('Pearson Coefficient is', pearsonCoef)
```
